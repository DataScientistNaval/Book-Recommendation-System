```python
books.drop_duplicates(inplace=True)
users.drop_duplicates(inplace=True)
ratings.drop_duplicates(inplace=True)
```

```python
df=ratings.merge(books, on='ISBN')
```

```python
df.drop(['Image-URL-S','Image-URL-M','Image-URL-L'],axis=1,inplace=True)
```

```python
df
```

```python
users=df.groupby('User-ID').count()['Book-Rating']>200
```

```python
users
```

```python
finalusers=users[users].index
finalusers
```

```python
filtered_user_rating=df[df['User-ID'].isin(finalusers)]
filtered_user_rating.head()
```

```python
books=filtered_user_rating.groupby('Book-Title').count()['Book-Rating']>50
```

```python
books
```

```python
finalbooks=books[books].index
finalbooks
```

```python
finaldf=filtered_user_rating[filtered_user_rating['Book-Title'].isin(finalbooks)]
```

```python
pivot=finaldf.pivot_table(index='Book-Title',values='Book-Rating',columns='User-ID')
```

```python
pivot
```

```python
pivot.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pivot)
```

```python
def recommend(Book):
    indexfetch=np.where(pivot.index==Book)[0][0]
    distance=sorted(list(enumerate(similarity_score[indexfetch])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in distance:
        print(pivot.index[i[0]])
```

```python
recommend('You Belong To Me')
```
